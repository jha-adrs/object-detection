```python
import time
import os
from sagemaker import get_execution_role, session
import boto3

region = boto3.Session().region_name
role = get_execution_role()
sm_client = boto3.client('sagemaker', region_name=region)
```

```python
# Creating the model group
import time
model_package_group_name = f"model-group-{time.strftime('%Y-%m-%d-%H-%M-%S', time.gmtime())}"
model_package_group_input_dict = {
    "ModelPackageGroupName": model_package_group_name,
    "ModelPackageGroupDescription": "Object Detection Model Groups",
}

create_model_package_group_response = sm_client.create_model_package_group(**model_package_group_input_dict)
print('ModelPackageGroup Arn : {}'.format(create_model_package_group_response['ModelPackageGroupArn']))
```

```python
from sagemaker import image_uris

image_uris.retrieve("object-detection", region)
```

```python
# TO Invoke the endpoint (Serverless)
import boto3
runtime = boto3.client(
    'sagemaker-runtime', 
    region_name='us-east-1',
    aws_access_key_id='',
    aws_secret_access_key=''
    )
```

```python
endpoint_name = "Tensorflow-OD"
content_type = 'image/jpeg'
payload = open('test.jpg', 'rb').read()
response = runtime.invoke_endpoint(
    EndpointName=endpoint_name,
    ContentType=content_type,
    Body=payload
)
```
